**1. Escribir un programa que reciba una lista de números enteros y determine
la cantidad de números pares e impares**

```python
def coun_pares_e_impares(numero):
  while True:
    try:
      numeros_listados = [int(x) for x in numero.split(',')]
      pares = 0
      impares = 0
      for n in numeros_listados:
        if n % 2 == 0:
            pares += 1
        else:
            impares += 1
      return pares, impares
    except ValueError:
            print("Entrada inválida. Asegúrate de ingresar solo números enteros separados por comas.")
    else:
      break


numeros=str(input("Escribe una lista de numeros separalos por coma ',': "))
pares, impares = coun_pares_e_impares(numeros)

print(f"Cantidad de números pares: {pares}")
print(f"Cantidad de números impares: {impares}")
```

# **Gráficos básicos con Matplotlib**

***Crear un gráfico de barras que agrupe las ventas por sucursal y mes, y sume las cantidades vendidas.:***

```python
# Agrupar datos por sucursal y mes, sumando las cantidades vendidas
sales_by_branch_month = data.groupby(['Sucursal', 'Mes'])['Cantidad_vendida'].sum().reset_index()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Mes', y='Cantidad_vendida', hue='Sucursal', data=sales_by_branch_month)
plt.title('Ventas por Sucursal y Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Crear un gráfico de dispersión para mostrar la relación entre las columnas “Precio Unitario” y “Cantidad Vendida”**

```python

y=data['Precio_unitario']
x=data['Cantidad_vendida']
colors = ['red' if precio < 20 else 'blue' for precio in y]
xmin, xmax = x.min(), x.max()
ymin, ymax = y.min(), y.max()
plt.plot([xmin, xmax], [ymin, ymax], color='purple', linestyle='--', linewidth=2, label='Línea diagonal')
plt.scatter(x,y,c=colors)
plt.title('Precio Unitario vs Cantidad Vendida')
plt.xlabel('Precio Unitario')
plt.ylabel('Cantidad Vendida')
plt.show()

```
